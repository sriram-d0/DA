```python
#a)	How to identify the unique values in the region column.
pd.Series(ava["region"].unique())
```

```python
#b)	What is the maximum price for an avocado in the dataset.
ava["AveragePrice"].max()
```

```python
#c)	Identify the type distribution  and take a single avocado in the dataset and find out the median price ,mean, and standard deviation.
print(ava["type"].value_counts())
single=ava[ava["type"]=='conventional']
print(single["AveragePrice"].mean())
print(single["AveragePrice"].median())
print(single["AveragePrice"].std())
```

```python
#d)	Find the highest, lowest price for conventional avocadoâ€™s in year with location.
conv=ava[ava["type"]=="conventional"]
p=conv.groupby(['year', 'region'])["AveragePrice"]
print(p.max())
print(p.min())
```

```python
#e)	Draw the plots of the distribution of average price for different types of Avocados
import seaborn as sns
sns.boxplot(x='type', y='AveragePrice', data=ava)
```

```python
#f)	Find the correlation matrix to measure the strength of the correlation between variables.
ava.select_dtypes(include=['float64', 'int64']).corr()
```

```python
#g)	Find out the volume of avocado sales has increased in the last 5 years.
current_year = ava['year'].max()
five_years_ago = current_year - 5

last_5_years_volume = ava[ava['year'] >= five_years_ago]['Total Volume'].sum()
total_volume = ava['Total Volume'].sum()
last_5_years_volume - total_volume
```
