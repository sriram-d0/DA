###Series

```python
import pandas as pd
import numpy as np

# a) Operations on two pandas Series
series1 = pd.Series([2, 4, 6, 8, 10])
series2 = pd.Series([1, 3, 5, 7, 10])

addition = series1 + series2
subtraction = series1 - series2
multiplication = series1 * series2
division = series1 / series2

print("Addition:")
print(addition)
print("Subtraction:")
print(subtraction)
print("Multiplication:")
print(multiplication)
print("Division:")
print(division)
```

```python
# b) String operations on a given pandas series
string_series = pd.Series(['Hello', 'World', 'Python', 'Pandas'])
upper_case = string_series.str.upper()
lower_case = string_series.str.lower()
string_length = string_series.str.len()
print("Uppercase:")
print(upper_case)
print("Lowercase:")
print(lower_case)
print("String Length:")
print(string_length)
```

```python
# c) Remove whitespaces from string values
whitespace_series = pd.Series(['   Data', 'Science   ', '  Machine Learning  '])
remove_whitespace = whitespace_series.str.strip()
remove_left_whitespace = whitespace_series.str.lstrip()
remove_right_whitespace = whitespace_series.str.rstrip()

print("Remove Whitespace:")
print(remove_whitespace)
print("Remove Left Whitespace:")
print(remove_left_whitespace)
print("Remove Right Whitespace:")
print(remove_right_whitespace)
```

```python
# d) Create a series from a list, numpy array, and dictionary
list_series = pd.Series([1, 2, 3, 4, 5])
array_series = pd.Series(np.array([10, 20, 30, 40, 50]))
dict_series = pd.Series({'a': 100, 'b': 200, 'c': 300, 'd': 400, 'e': 500})
print("List Series:",list_series)
print("Array Series:",array_series)
print("Dictionary Series:",dict_series)
```

```python
# e) Calculate the number of characters in each word in a series
word_series = pd.Series(['apple', 'banana', 'cherry', 'date'])
word_length = word_series.str.len()
print("Word Length:")
print(word_length)
```

```python
# f) Compare two Pandas Series
comparison_result = (series1 == series2)
print("Comparison Result:")
print(comparison_result)
```

```python
# g) Convert a Pandas Series to a Python list and its type
list_from_series = series1.tolist()
type_of_series = type(series1)
print("List from Series:")
print(list_from_series)
print(type_of_series)
```

```python
# i) Combine many series to form a dataframe
data = {'Series1': series1, 'Series2': series2, 'ListSeries': list_series}
df = pd.DataFrame(data)
print("DataFrame:")
print(df)
```

```python
# j) Stack two series vertically and horizontally
stacked_vertical = pd.concat([series1, series2], axis=0)
stacked_horizontal = pd.concat([series1, series2], axis=1)
print("Stacked Horizontally:")
print(stacked_horizontal)
print("Stacked Vertically:")
print(stacked_vertical)
```

```python
# k) Create and display a DataFrame from a specified dictionary data with index labels
data_dict = {'Name': ['Alice', 'Bob', 'Charlie'], 'Age': [25, 30, 35]}
index_labels = ['a', 'b', 'c']
df_with_index = pd.DataFrame(data_dict, index=index_labels)
print("DataFrame with Index:")
print(df_with_index)
```

```python
# l) Identify frequency counts of unique items in a series
fruit_series = pd.Series(['apple', 'banana', 'apple', 'cherry', 'banana', 'apple'])
item_counts = fruit_series.value_counts()
print("Item Counts:")
print(item_counts)
```

```python
# m) Get the items of series A not present in series B
series_a = pd.Series([1, 2, 3, 4, 5])
series_b = pd.Series([3, 4, 5, 6, 7])
items_not_in_b = series_a[~series_a.isin(series_b)]
print("Items not in Series B:")
print(items_not_in_b)
```

```python
# n) Convert a numpy array to a DataFrame of a given shape
import numpy as np

numpy_array = np.array([[1, 2], [3, 4], [5, 6]])
shape_df = pd.DataFrame(numpy_array, columns=['A', 'B'],index=['A','B','C'])
print("DataFrame with Shape:")
print(shape_df)
```
